```python
#Example:

my_dogs = 2

my_dogs = ["Sammy", "Frankie"]
```

##### Pro's of Dynamic Typing:
1. we don't need to declare the data type.
2. Faster Development time.


##### Con's of Dynamic Typing:

2. should be aware of type() function

##### Slicing a String:

```python
#Take a String

myString = 'asdfghjikl'
```

```python
myString
```

```python
#Start from a particular sub-section of the String ex. from index 2 (Value: c) until the end of the string

myString[2:] #Start index is set to 2, which means 3 chars are removed from the string
```

```python
myString[:3] #StopIndex is set to index 3 which means start from index 0 and move until index 3
```

```python
myString[3:6] #Trying to get the chars between index of 3 and 6 -> Start index is 3 and stop index is 6
```

```python
myString[1:3] #Get strings between index 1 until 3
```

```python


myString[::] 

#Get all the chars from index 0 to last index
```

```python
#Step Size: Jump n chars in a string

myString[::2] #Go in Jumps of 2 
```

```python
myString[::3]
```

```python
myString[2:7:2]

#Start Index: 2 | Stop Index: 7 | Step Size: 2
```

```python
myString[::-1]

#String Reversal
```

```python
myString[::-2]

#Stepsize in reverse direction
```

```python
myString[6:1:-1]

#In reverse order, start from index 6 and stop at index 1 and jump size is -1 
```
