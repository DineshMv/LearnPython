#### Comparision Operators

An Example of comparision operators are shown below:



##### **Equality Operator:**   **Syntax: ==**

```python
#Integer comparision

2==2
```

```python
#Here we are trying to assign value of 2 to a var 2.. which is an error
2=2
```

```python
2==1
```

```python
'hello' == 'bye'
```

```python
#String 2 has a H, so both are not equal
'hello' == 'Hello'
```

```python
#String comparision

'hello' == 'hello'
```

```python
#String to int comparision
'2' == 2
```

```python
2.0 == 2
```
